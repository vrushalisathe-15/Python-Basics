```python
"""Find the sum of first 10 natural numbers using while loop.
Hint: Use sum += i inside the loop."""
sum=0
i=1
while i<=10:
    sum+=i
    i+=1
print(sum)
```

```python
""" Find factorial of a number using while loop.
Hint: Use fact *= num and decrement num."""
num=int(input("Enter the number"))
fact=1
while num > 0:
    fact*=num
    num-=1
print(fact)
```

```python
""" Count the number of digits in a given number using while loop.
Hint: Use num // 10 and count iterations."""
num = int(input("Enter the number"))
count = 0
while num != 0:
    num = num // 10
    count += 1
print(count)

```

```python
"""Reverse a given number using while loop.
Hint: Extract last digit using num % 10 and form reverse."""
num = int(input("Enter a number: "))
reverse = 0

while num != 0:
    digit = num % 10          # Get last digit
    reverse = reverse * 10 + digit
    num = num // 10           # Remove last digit

print("Reversed number:", reverse)

```

```python
""" Check if a number is palindrome using while loop.
Hint: Compare the reversed number with original."""
num = int(input("Enter a number: "))
original= num
reverse = 0

while num != 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if original == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome number")

```

```python
#using Slicing
num=input("enter the value")
rev=num[::-1]
if num == rev:
    print("Palindrome")
else:
    print("Not palindrome")
```
