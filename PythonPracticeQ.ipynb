```python
#While Loop
```

```python
"""Print numbers from 1 to 10"""
num=0
while num<=10:
    print(num)
    num+=1
```

```python
"""Reverse a number."""
num=1234
rev=0
while num>0:
    rev=rev*10+num%10
    num//=10
print(rev)
  
```

```python
"""Sum of digits of a number."""
num=10
s=0
while num>0:
    digit=num%10
    s=s+digit
    num
```

```python
"""Check palindrome number"""
num = int(input("Enter the number: "))
original= num
reverse = 0
while num != 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if original == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

```python
"""Find factorial of a number."""
num=int(input("Enter the number"))
fact=1
while num > 0:
    fact*=num
    num-=1
print(fact)
```
