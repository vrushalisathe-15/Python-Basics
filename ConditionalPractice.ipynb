```python
"""Write a Python program that checks if a number is greater than
100 and less than 200 using an if statement."""
num=int(input("Enter the number"))
if num>=100 & num<=200: # ALSO 100<num<200
    print("Number is in between 100 & 200")
else:
    print("Not in between 100 & 200")
```

```python
""" Write a Python program that checks if a given year is a leap year using an if statement. 
(Hint: A leap year is divisible by 4 but not by 100 unless also
divisible by 400.)"""
year=int(input("Enter year"))
if (year%4==0 and year%4!=0) or (year % 400 == 0):
    print("Year is leap")
else:
    print("Year is not leap")
```

```python
""" Write a Python program that takes an input age and prints whether the
person is a child (0-12), teenager (13-19), adult (20-64), or senior (65+)."""
age=int(input("Enter age"))
if 0<age<12:
    print("Child")
elif 13<age<19:
    print("Teenager")
elif 20<age<64:
    print("Adult") 
else:
    print("Senior")
```

```python
"""Write a program that checks whether a number is divisible by both 3
and 5 using logical operators."""
num=int(input("Enter the number"))
if num % 3 == 0:
    print("Number is divisible by 3 ")
elif num % 5 == 0:
    print("Number is divisible by  5")
else:
    print("Not divisible by 3 & 5")
```

```python
""" Create a program that asks for the user’s score and prints the
corresponding grade:
-If the score is 90 or above, print "A"
-If the score is between 80 and 89, print "B"
-If the score is between 70 and 79, print "C"
-If the score is below 70, print "F" """
score=int(input("Enter your score"))
if score>=90:
    print("A")
elif 80<score<89:
    print("B")
elif 70<score<79:
    print("C")
else:
    print("F")
```
