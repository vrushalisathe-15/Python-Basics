```python
#For Loop with Strings
"""Print each character of text one by one."""
text="Hello World! Python is Amazing"
for i in text:
    print(i,end=" ")
```

```python
"""Count the number of vowels in text."""
text="Hello World! Python is Amazing"
vowel="aeiouAEIOU"
count=0
for i in text:
    if i in vowel:
        count+=1
print(count)        
```

```python

text="Hello World! Python is Amazing"
vowel="aeiouAEIOU"
count=0
for i in text:
    if i in vowel:
        count+=1
    else:
        count+=0
print(count)
```

```python
"""Extract only the uppercase letters from text."""
text="Hello World! Python is Amazing"
for i in text:
    if i.isupper():
      print(i)
```

```python
"""Reverse text using a for loop."""
text="Hello World! Python is Amazing"
rev=""
for i in range(len(text)-1,-1,-1):
    rev+=text[i]
print(rev)
```

```python
"""Count how many transactions are above $1000 from transactions"""
transactions = [500, 1200, 300, 4000, 750, 980, 2500]
count = len(list(filter(lambda t: t > 1000, transactions)))
print("Transactions above $1000:", count)
```

```python
"""Calculate the percentage increase or decrease in stock_prices compared to the
previous day. """
stock_prices = [101, 105, 99, 110, 120, 98, 115]
for i in range(1, len(stock_prices)):
    prev = stock_prices[i - 1]
    cur = stock_prices[i]

    percentage_change = ((cur - prev) / prev) * 100 #calculate percentage change

    print(f"Day {i}: {percentage_change:.2f}%")  # - 
```

```python
""" Print all employee names from employee_salaries."""
employee_salaries = {"John": 50000, "Alice": 75000, "Bob": 60000, "Mike": 90000}
for n in employee_salaries:
    print(n)
```

```python
"""Increase each employee's salary in employee_salaries by 10% and print the
new salaries."""
employee_salaries = {"John": 50000, "Alice": 75000, "Bob": 60000, "Mike": 90000}
for employee in employee_salaries:
    employee_salaries[employee] *= 1.10

# Print new salaries
for employee, salary in employee_salaries.items():
    print(employee, ":", int(salary))
```

```python
"""Count students who scored above 80 in student_scores."""
student_scores = {"Emma": 85, "Oliver": 78, "Sophia": 92, "Liam": 88}
count=0
for i in student_scores.values():
    if i > 80:
        count+=1
print(count)
```

```python
"""Find and print the customers with negative balances from bank_accounts."""
bank_accounts = {"John": 1200, "Alice": -500, "Bob": 3000, "Mike": -200}
for cust , bal  in bank_accounts.items():
    if bal < 0:
        print(cust)
```

```python
""" Print only the out-of-stock products from inventory."""
inventory = {"Laptop": 10, "Mouse": 25, "Keyboard": 30, "Monitor": 5, "Printer": 0}
for product , stock  in inventory.items():
    if stock == 0:
        print(product)
```
