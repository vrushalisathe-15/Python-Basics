```python
"""Write a program that counts the number of occurrences of each word in a sentence using a dictionary.""" 
sentence = "python is easy and python is powerful"

words = sentence.split() #convert into list
word_count = {}  #dictonary will store word â†’ number of times it appears 

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(word_count)
```

```python
"""Create a dictionary of five students with their names as keys and ages as values. Print the
student with the maximum age."""
student={"Vrushali":20,
         "Gargi":21,
         "Shifa":25,
         "Sakshi":19,
         "Shweta":24
        }
max_age=max(student,key=student.get)
print("Student with maximum age:",max_age)
print("Age:",student[max_age])
```

```python
"""Given a dictionary storing movie names and their ratings, sort it in descending order of ratings
and print the result"""
movies = {
    "Inception": 8.8,
    "Interstellar": 8.6,
    "Titanic": 7.8,
    "Avatar": 7.9
}
movies
```

```python
sorted_movies = dict(sorted(movies.items(), key=lambda x: x[1], reverse=True))
print(sorted_movies)
```

```python
"""Use both tuples and dictionaries together:
 - Create a dictionary where keys are countries and values are tuples of (capital, population,
currency).
 - Print the capital of 'Japan' and the currency of 'France'."""
countries = {
    "India": ("New Delhi", "1.4 billion", "INR"),
    "Japan": ("Tokyo", "126 million", "Yen"),
    "France": ("Paris", "67 million", "Euro")
}
print("Capital of Japan:",countries["Japan"][0])
print("Currency of France:",countries["France"][2])
```

```python
"""Complex Nested Dictionary & Tuple:
 - Store employee records where keys are employee names and values are tuples containing (age,
department, salary).
 - Find the average salary of all employees in a given department."""
employees={
    "vrushali":(20,"IT",50000),
    "Gargi":(21,"IT",51000),
    "sakshi":(22,"HR",25000)
}
total=0
count=0
for emp in employees.values():
    if emp[1] == "IT":
        total += emp[2]
        count += 1

print(total / count)
```
