```python
"""Given text = "apple,banana,grape,orange",
 - Split the string into a list.
 - Extract the second element from the list and capitalize it.
 - Join the list back into a string using " | " as a separator"""
text = "apple,banana,grape,orange"
b=text.split(",")
```
