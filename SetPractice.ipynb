```python
#Create a set with the elements {1, 2, 3, 4, 5} and print it.
set1={1,2,3,4,5}
set1
```

```python
#Add an element 6 to the set {1, 2, 3, 4, 5}
set1.add(6)
set1
```

```python
#Remove the element 3 from the set {1, 2, 3, 4, 5}
set1.remove(3)
set1
```

```python
#Check if 4 exists in the set {1, 2, 3, 4, 5}.
4 in set1
```

```python
#Create a set from the list [1, 2, 2, 3, 4, 4, 5] and print it
list=[1,2,2,3,4,4,5]
set(list)
```

```python
####Intermediate Level
```

```python
#Find the union of sets {1, 2, 3} and {3, 4, 5}
set1={1,2,3}
set2={3,4,5}
set1 | set2
```

```python
# Find the intersection of sets {1, 2, 3} and {2, 3, 4}.
set1={1,2,3}
set2={2,3,4}
set1 & set2
```

```python
#Find the difference between sets {1, 2, 3, 4} and {3, 4, 5, 6}
set1={1,2,3,4}
set2={3,4,5,6}
set1 - set2
```

```python
#Find the symmetric difference between sets {1, 2, 3} and {3, 4, 5}
set1={1,2,3}
set2={3,4,5}
set1 ^ set2
```

```python
#Check if {1, 2} is a subset of {1, 2, 3, 4}.
set1={1,2}
set2={1,2,3,4}
set1.issubset(set2)
```

```python
#######  ADVANCE LEVEL
```

```python
# Remove all duplicates from the string 'programming' using a set.
text="programming"
new=set(text)
res="".join(new)
res
```

```python
#Find common elements between two lists using sets:
# list1 = [1, 2, 3, 4, 5]
 #list2 = [4, 5, 6, 7, 8] 
list1=[1,2,3,4,5]
list2=[4,5,6,7,8]
set(list1) & set(list2)
```

```python
#Find elements that are in list1 but not in list2 using sets.
set(list1) - set(list2)
```

```python
#Check if two sets are disjoint (no common elements)
set1={1,2,3}
set2={4,5,6}
set1.isdisjoint(set2)
```

```python
#Convert a tuple (1, 2, 3, 4) into a set.
tuple=(1,2,3,4)
set(tuple)
```
