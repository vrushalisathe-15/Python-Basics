```python

correct_password = "Secure@123"
attempts = 0
while True:
    password = input("Enter the Password")
    
    if password == correct_password:
        print("Correct Password. Access granted")
        break
    else:
        attempts += 1
        print("Wrong password")
    if attempts == 3:
        print("TOO many attempts!!! Try Again Later...")
        break
```
